## Use case

AO is generally used to affirm trends or to anticipate possible reversals.

```python
import pandas_ta as ta
import pandas as pd
import yfinance as yf
import mplfinance as mpf
```

```python
df = pd.DataFrame().ta.ticker("TSLA")
start_date = '2023-01-01'
df = df[df.index >= start_date]
df
```

```python

df['hlavg'] = df.apply(lambda row: 0.5 * (row['High'] + row['Low']), axis=1)
df['sma5'] = df['hlavg'].rolling(window=5).mean()
df['sma34'] = df['hlavg'].rolling(window=34).mean()
df['ao'] = df['sma5'] - df['sma34']
df = df.drop(columns=['hlavg', 'sma5', 'sma34'])
df

```

```python
df.ta.ao()
```

```python
apds = [
    mpf.make_addplot(df['ao'], panel=1, color='g', secondary_y=False),  # Indicator below candlestick
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```

```python
apds = [
    mpf.make_addplot(df.ta.ao(), panel=1, color='g', secondary_y=False),  # Indicator below candlestick
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```

# ROC

Shows momentum

Can indicate if a certain instrument is oversold or overbought if it reaches a certain threshold



```python
n = 10
df['ROC'] = (df['Close'] - df['Close'].shift(n)) / df['Close'].shift(n) * 100

apds = [
    mpf.make_addplot(df['ROC'], panel=1, color='g', secondary_y=False)
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```

```python
apds = [
    mpf.make_addplot(df.ta.roc(length=10), panel=1, color='g', secondary_y=False)
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```

# Momentum



```python
n = 50
df['Momentum'] = df['Close'] - df['Close'].shift(n)
```

```python
apds = [
    mpf.make_addplot(df['Momentum'], panel=1, color='g', secondary_y=False)
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```

```python
apds = [
    mpf.make_addplot(df.ta.mom(length=50), panel=1, color='g', secondary_y=False)
]

mpf.plot(df, type='candle', addplot=apds, volume=False)
```
